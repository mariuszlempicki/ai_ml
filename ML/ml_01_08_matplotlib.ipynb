```python
n = 200
x = np.linspace(0, 10, n)
y = x**2 + np.random.uniform(0, 10, n)
z = 0.5*x**2 + np.random.uniform(0, 10 , n)
```

```python
plt.figure(figsize=(15,5))
plt.plot(x, y, color="red", linewidth=1, linestyle="--", label="y")
plt.plot(x, z, label="z")
plt.title("Wykres liniowy")
plt.xlabel("X")
plt.xlim(1, 9)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.scatter(x, y, color="orange", s=10, marker="D")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.hist(y, bins=3)
plt.show()
```

```python
plt.hist(y, bins=[0, 50, 100, 150])
plt.show()
```
