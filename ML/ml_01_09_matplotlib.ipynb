```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 200
x = np.linspace(0, 10, n)
y = x**2 + np.random.uniform(0, 10, n)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1, 3, 1)
plt.scatter(x, y)

plt.subplot(1, 3, 2)
plt.plot(x, y)

plt.subplot(1, 3, 3)
plt.hist(y)

plt.show()
```
