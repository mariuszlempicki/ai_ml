## Rysowanie wykresów

1. Narysuj wykres liniowy funkcji $f(x) = x^2$ dla $x$ należącego do przedziału [-10, 10].

---
(czas: 6 min.)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definiujemy funkcję, np. f(x) = x^2
def f(x):
    return x**2

# Tworzymy wartości z przedziału [-10, 10]
x = np.linspace(-10, 10, 400)
y = f(x)

# Rysujemy wykres
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="f(x) = x^2", color="blue")

# Dodajemy tytuł i etykiety osi
plt.title("Wykres funkcji f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")

# Dodajemy legendę i siatkę
plt.legend()
plt.grid(True)

# Wyświetlamy wykres
plt.show()
```

2. Narysuj na jednym wykresie wykresy liniowe funkcji $f(x) = \text{sin}(x)$ oraz $f(x) = \text{cos}(x)$

(czas: 6 min.)

```python
# Tworzymy wartości x z przedziału [-10, 10]
x = np.linspace(-10, 10, 400)

# Definiujemy funkcje f(x) = sin(x) oraz f(x) = cos(x)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Rysujemy wykresy
plt.figure(figsize=(8, 6))
plt.plot(x, y_sin, label="f(x) = sin(x)", color="blue")
plt.plot(x, y_cos, label="f(x) = cos(x)", color="green")

# Dodajemy tytuł i etykiety osi
plt.title("Wykresy funkcji f(x) = sin(x) oraz f(x) = cos(x)")
plt.xlabel("x")
plt.ylabel("f(x)")

# Dodajemy legendę i siatkę
plt.legend()
plt.grid(True)

# Wyświetlamy wykres
plt.show()
```

3. Oba powyższe wykresy zmodyfikuj tak aby były wykresami punktowymi.

---
(czas: 2 min.)

```python
# Tworzymy wartości x z przedziału [-10, 10]
x = np.linspace(-10, 10, 400)

# Definiujemy funkcje f(x) = sin(x) oraz f(x) = cos(x)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Rysujemy wykresy punktowe
plt.figure(figsize=(8, 6))
plt.plot(x, y_sin, 'o', label="f(x) = sin(x)", color="blue")  # 'o' oznacza punkty
plt.plot(x, y_cos, 'x', label="f(x) = cos(x)", color="green")  # 'x' oznacza punkty

# Dodajemy tytuł i etykiety osi
plt.title("Wykresy punktowe funkcji f(x) = sin(x) oraz f(x) = cos(x)")
plt.xlabel("x")
plt.ylabel("f(x)")

# Dodajemy legendę i siatkę
plt.legend()
plt.grid(True)

# Wyświetlamy wykres
plt.show()
```

4. Wczytaj dane z pliku `weight-height.csv`. Narysuj wykres punktowy obrazujący zależność między wagą a wzrostem ludzi. Następnie podziel zbiór danych na dwie części (na podstawie płci) i stwórz dwa analogiczne wykresy, które będą na siebie nałożone.

---

(czas: 9 min.)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Wczytujemy dane z pliku CSV
data = pd.read_csv('data\weight-height.csv')

# Tworzymy wykres punktowy obrazujący zależność między wagą a wzrostem dla wszystkich danych
plt.figure(figsize=(10, 6))
plt.scatter(data['Height'], data['Weight'], alpha=0.5, label='All data', color='gray')

# Dodajemy tytuł i etykiety osi
plt.title('Zależność między wagą a wzrostem')
plt.xlabel('Wzrost (cale)')
plt.ylabel('Waga (funty)')
plt.legend()
plt.grid(True)
plt.show()

# Tworzymy dwa zbiory danych na podstawie płci
data_male = data[data['Gender'] == 'Male']
data_female = data[data['Gender'] == 'Female']

# Tworzymy wykres punktowy dla mężczyzn i kobiet
plt.figure(figsize=(10, 6))
plt.scatter(data_male['Height'], data_male['Weight'], alpha=0.5, label='Mężczyźni', color='blue')
plt.scatter(data_female['Height'], data_female['Weight'], alpha=0.5, label='Kobiety', color='red')

# Dodajemy tytuł i etykiety osi
plt.title('Zależność między wagą a wzrostem z podziałem na płeć')
plt.xlabel('Wzrost (cale)')
plt.ylabel('Waga (funty)')

# Dodajemy legendę i siatkę
plt.legend()
plt.grid(True)

# Wyświetlamy wykres
plt.show()
```

5. Wczytaj dane z pliku `AirPassengers.csv` i stwórz wykres liniowy liczby pasażerów na przestrzeni czasu.

---
(czas: 5 min.)

```python
# Wczytujemy dane z pliku CSV
data = pd.read_csv('data\AirPassengers.csv')

# Konwertujemy kolumnę 'Month' na typ daty
data['Month'] = pd.to_datetime(data['Month'], format='%Y-%m')

# Ustawiamy kolumnę 'Month' jako indeks (dla lepszej prezentacji na wykresie)
data.set_index('Month', inplace=True)

# Tworzymy wykres liniowy liczby pasażerów na przestrzeni czasu
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['#Passengers'], label='Liczba pasażerów', color='blue')

# Dodajemy tytuł i etykiety osi
plt.title('Liczba pasażerów na przestrzeni czasu')
plt.xlabel('Rok')
plt.ylabel('Liczba pasażerów')

# Dodajemy siatkę i legendę
plt.grid(True)
plt.legend()

# Wyświetlamy wykres
plt.show()
```

6. Wczytaj dane z pliku `exchange_rates.csv` i stwórz wykres liniowy, na którym zobrazujesz zmiany ceny dolara w poszczególnych walutach.

---
(czas: 8 min.)

```python
# Wczytujemy dane z pliku CSV
data = pd.read_csv('data\exchange_rates.csv', header=None)
print(data)
# Konwertujemy kolumnę 'timestamp' na format daty
data['timestamp'] = pd.to_datetime(data['timestamp'])

# Ustawiamy kolumnę 'timestamp' jako indeks (dla lepszej prezentacji na wykresie)
data.set_index('timestamp', inplace=True)

# Tworzymy wykres liniowy zmiany kursu dolara względem różnych walut
plt.figure(figsize=(12, 8))

# Rysujemy wykresy dla poszczególnych walut względem USD
plt.plot(data.index, data['USD-EUR'], label='USD-EUR', color='blue')
plt.plot(data.index, data['USD-GBP'], label='USD-GBP', color='green')
plt.plot(data.index, data['USD-CHF'], label='USD-CHF', color='red')
plt.plot(data.index, data['USD-PLN'], label='USD-PLN', color='orange')

# Dodajemy tytuł i etykiety osi
plt.title('Zmiany ceny dolara (USD) w poszczególnych walutach')
plt.xlabel('Czas')
plt.ylabel('Kurs dolara (USD)')

# Dodajemy siatkę, legendę i formatujemy oś czasu
plt.grid(True)
plt.legend()

# Wyświetlamy wykres
plt.show()
```

7. Wygeneruj wektor 1000 losowych wartości z rozkładu normalnego standardowego (funkcja `np.random.randn`). Następnie przedstaw je na histogramie. Zobacz jak zmiana liczby słupków wpływa na wygląd wykresu.

---
(czas: 8 min.)

```python
# Generowanie 1000 losowych wartości z rozkładu normalnego standardowego
data = np.random.randn(1000)

# Tworzenie histogramu
plt.figure(figsize=(12, 6))

# Histogram z różnymi liczbami słupków
bins_list = [10, 20, 50, 100]

for i, bins in enumerate(bins_list):
    plt.subplot(2, 2, i + 1)  # Ustawienie subwykresów
    plt.hist(data, bins=bins, color='blue', alpha=0.7, edgecolor='black')
    plt.title(f'Histogram z {bins} słupkami')
    plt.xlabel('Wartości')
    plt.ylabel('Częstość')

plt.tight_layout()  # Dostosowanie układu
plt.show()
```

8. Wczytaj plik `cars.csv`. Stwórz histogramy ceny, mocy oraz pojemności silnika samochodów. Jeżeli będzie potrzeba, przefiltruj zbiór danych, odsiewając zbyt duże wartości.

---
(czas: 10 min.)

```python
# Wczytywanie pliku CSV
data = pd.read_csv('data\cars.csv', header=None)

# Kolumny
price = data[0]
power = data[7]
engine_vol = data[4]
print(price)
print(power)
print(engine_vol)
```

```python
# Filtracja wartości (przykładowe wartości graniczne)
price_filtered = price[price < 100000]  # Przykład: filtrujemy ceny powyżej 100000
power_filtered = power[(power > 0) & (power < 500)]  # Przykład: filtrujemy moc powyżej 0 i poniżej 500
engine_vol_filtered = engine_vol[(engine_vol > 0) & (engine_vol < 10)]  # Przykład: filtrujemy pojemność poniżej 10
```

```python
price_filtered = price
power_filtered = power
engine_vol_filtered = engine_vol
```

```python
# Tworzenie histogramów
plt.figure(figsize=(15, 5))

# Histogram ceny
plt.subplot(1, 3, 1)
plt.hist(price_filtered, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Histogram cen samochodów')
plt.xlabel('Cena')
plt.ylabel('Częstość')

# Histogram mocy
plt.subplot(1, 3, 2)
plt.hist(power_filtered, bins=30, color='green', alpha=0.7, edgecolor='black')
plt.title('Histogram mocy samochodów')
plt.xlabel('Moc (KM)')
plt.ylabel('Częstość')

# Histogram pojemności silnika
plt.subplot(1, 3, 3)
plt.hist(engine_vol_filtered, bins=30, color='orange', alpha=0.7, edgecolor='black')
plt.title('Histogram pojemności silnika')
plt.xlabel('Pojemność (L)')
plt.ylabel('Częstość')

plt.tight_layout()  # Dostosowanie układu
plt.show()
```
